# Przykłady użycia wykresu heksagonalnego w analizie danych

# Przykład 1

```python
#importujemy niezbędne pakiety
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import matplotlib.cm as cm
import cmocean
import random

#generujemy dane (x,y)
#polecenie random.randn() zwraca próbkę rozkładu normalnego
#np.random.seed(2)
x = np.random.randn(5000)
y = np.random.randn(5000)

#tworzymy obszar rysowania
fig, ax = plt.subplots()

#tworzymy wykres heksagonalny gdzie x oraz y są osiami wykresu, gridsize jest rozmiarem punktów
#(większy gridsize to więcej punktów/sześciokątów, jest większa dokładność heksagonów)
#cmap jest wybraną przez nas mapą kolorów 
im = ax.hexbin(x, y, gridsize=30, cmap=cmocean.cm.ice)

#ustawiamy rozmiary osi x oraz y
ax.set(xlim=(-3, 3), ylim=(-3, 3))

#colorbar czyli paski kolorów to wizualizacja odwzorowania wartości skalarnych na kolory.
#W Matplotlib są one rysowane na dedykowanej osi
#add_axes([x, y, grubość, wysokość])
cbar_ax=fig.add_axes([0.95, 0.12, 0.04, 0.7])
fig.colorbar(im, cax=cbar_ax)

#Pokazujemy wykres
plt.show()
```

```python
#Importujemy niezbędne pakiety
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import cmocean

#Tworzymy wykres typu hexbin na podstawie danych z pliku csv
#Jako oś X dajemy kolumnę "iq", a jako oś Y dajmy kolumnę "cgpa" - czyli średnią studencką
#Ustawiamy minimalnego hexagona na wartość 1
#Ustawiamy blendowanie przezroczystości na 0.7
#Gridsize ustawiamy na 30, aby zwiększyć dokładność hexagonów
#Dodatkowo ustawiamy ładne kolorki :)
df.plot.hexbin(x="iq", y="cgpa", mincnt=1, alpha=0.7, gridsize=30, cmap=cmocean.cm.algae)


#Pokazujemy wykres
plt.show()
```

# Examples of using a hexagonal chart in data analysis

# Example 1

```python
#importing necessary packages
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import matplotlib.cm as cm
import cmocean
import random

#generate data (x,y)
#random.randn() returns a sample of the normal distribution
#np.random.seed(2)
x = np.random.randn(5000)
y = np.random.randn(5000)

#creating the drawing area
fig, ax = plt.subplots()

#creating a hexagonal graph where x and y are the axes of the graph and gridsize is the size of the points
#(greater gridsize means more points/hexagons, more hexagon accuracy)
#cmap is a color map of our choice 
im = ax.hexbin(x, y, gridsize=30, cmap=cmocean.cm.ice)

#setting the dimensions of the x and y axes
ax.set(xlim=(-3, 3), ylim=(-3, 3))

#colorbar is a visualization of the mapping of scalar values to colors.
#In Matplotlib they are drawn on a dedicated axis
#add_axes([x, y, thickness, height])
cbar_ax=fig.add_axes([0.95, 0.12, 0.04, 0.7])
fig.colorbar(im, cax=cbar_ax)

#showing plot
plt.show()
```

```python
#importing necessary packages
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.cm as cm
import cmocean

#Creating a hexbin chart based on data from a csv file
#As the x-axis we give the column "iq", and as the y-axis we give the column "cgpa" - i.e. the student average
#Setting the minimum hexagon to 1
#Setting the opacity blending to 0.7
#Setting gridsize to 30 to increase the accuracy of the hexagons
#Additionally, setting nice colors :)
df.plot.hexbin(x="iq", y="cgpa", mincnt=1, alpha=0.7, gridsize=30, cmap=cmocean.cm.algae)


#showing plot
plt.show()
```
